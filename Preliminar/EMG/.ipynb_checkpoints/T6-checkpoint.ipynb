## Prueba inicial

Se hace una prueba inicial con las caracteristicas originales, 

```python
import numpy as np
import scipy as sp
from sklearn.neural_network import MLPClassifier
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix


#se carga conjunto de entrenamiento y validacion
Train=np.load('Train.npy')
Val=np.load('Validation.npy') 


#Se separa cada conjunto en la señal y sus clases
sign_train = Train[:,0:6000];
class_train = Train[:,6000];

sign_val = Val[:,0:6000];
class_val = Val[:,6000];


#se caragan las imf previamente calculadas sobre los conjuntos de train y validacion
imf_tr=np.load('imf_tr.npy')
imf_val=np.load('imf_val.npy')            
#%%

#Se procede a extraer caracteristicas de las imf calculadas

#funciones de extraccion de caracteristicas entregadas
     
def crosses(my_array):
  return ((my_array[:,:-1] * my_array[:,1:]) < 0).sum(axis=1)

def slope_sign_changes(my_array):
    w = np.shape(my_array)[1];
    d1 = my_array[:,1:w-2] - my_array[:,2:w-1]
    d2 = my_array[:,2:w-1] - my_array[:,3:w]
    return np.double( np.sum((d1*d2) < 0, 1) );

def extr_feat(s):
    h = np.shape(s)[0]
    w = np.shape(s)[1];
    feat = np.zeros((h, 10))
    feat[:,0] = np.sum(np.abs(s),1); # iemg
    feat[:,1] = crosses(s); #zc
    feat[:,2] = slope_sign_changes(s); #ssc
    feat[:,3] = np.sum(np.abs(s[:,1:w-1] - s[:,2:w]),1); # wl
    feat[:,4] = np.sum(np.abs(s[:,1:w-1] - s[:,2:w] > 0.1),1); # wamp
    feat[:,5] = np.var(s,1);
    feat[:,6] = sp.stats.skew(s,1);
    feat[:,7] = sp.stats.kurtosis(s,1);
    feat[:,8] = np.median(s,1);
    feat[:,9] = np.std(s,1);
    return feat;

#se extraen las caracteristicas para Train y Validation
caract_train = np.hstack((extr_feat(sign_train[:,1:3000]),
                          extr_feat(sign_train[:,3000:6000]),
                          extr_feat(imf_tr[:,:,0]),
                          extr_feat(imf_tr[:,:,1]),
                          extr_feat(imf_tr[:,:,2]),
                          extr_feat(imf_tr[:,:,3]),
                          extr_feat(imf_tr[:,:,4]),
                          extr_feat(imf_tr[:,:,5])
                         ));

caract_val = np.hstack((extr_feat(sign_val[:,1:3000]),
                         extr_feat(sign_val[:,3000:6000]),
                         extr_feat(imf_val[:,:,0]),
                         extr_feat(imf_val[:,:,1]),
                         extr_feat(imf_val[:,:,2]),
                         extr_feat(imf_val[:,:,3]),
                         extr_feat(imf_val[:,:,4]),
                         extr_feat(imf_val[:,:,5])
                        ));






#%%

#se normalizan caracteristicas de cada conjunto, con los parametros del conjunto de train
scaler = preprocessing.StandardScaler().fit(caract_train);
caract_train = scaler.transform(caract_train);
caract_val = scaler.transform(caract_val);

#%%

#funcion para calcular desempeños
def TVFP(Confusion):
    #se crea arreglo con 0 para alberga TVP y TFP de cada clase
    TVFP=np.zeros((Confusion.shape[0],2))
    #suma sobre elementos de diagonal
    suma=0
    #Se itera sobre las clases de la matriz de confusion, par obtener TVP y TFP de cada
    #una
    for i in range(0, Confusion.shape[0]):
        #TVP se calcula dividiendo la diagonal de cada clase con el numero total
        #muestras de esa clase
        TVP= Confusion[i,i]/np.sum(Confusion[i,:])
        #TFP se calcula sumando los valores de la clase predicha de la clase y restandole el elemento diagonal,
        #luego esto se divide por todos los datos, menos la fila de la clase
        TFP=(np.sum(Confusion[:,i])-Confusion[i,i])/(np.sum(Confusion)-np.sum(Confusion[i,:]))
        TVFP[i,:]=[TVP,TFP]
        #suma diagonales
        suma=suma+Confusion[i,i]
    
    #Al final se agrega el promedio de TVP y TFP de la red   
    PromTVP=np.sum(TVFP[:,0])/TVFP.shape[0]
    PromTFP=np.sum(TVFP[:,1])/TVFP.shape[0]
    Prom=np.array([[PromTVP,PromTFP]])
    TVFP=np.concatenate((TVFP,Prom), axis=0)
    #se retorna el arreglo
    Acc=suma/np.sum(Confusion)
    return TVFP, Acc


#%%
#se entrena con MLP
clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300,50), max_iter = 10000, random_state=1);
clf.fit(caract_train, class_train);

#se clasifica conjunto de validacion
pred = clf.predict(caract_val);

#se calculan desempeños
conf = confusion_matrix( class_val, pred);
Rates, Acc= TVFP(conf)
print('MLP')
print('Acc= ', Acc*100 );
print('TVP= ', Rates[6,0]*100 );
print('TFP= ', Rates[6,1]*100 );
print( conf );
#%%

#se repite el proceso anterior pero para un clasificador Random Forest
clf = RandomForestClassifier(n_estimators=50,max_leaf_nodes=100,n_jobs=-1,random_state=0)
clf.fit(caract_train, class_train);
pred = clf.predict(caract_val);

conf = confusion_matrix( class_val, pred);
Rates, Acc= TVFP(conf)
print('RF')
print('Acc= ', Acc*100 );
print('TVP= ', Rates[6,0]*100 );
print('TFP= ', Rates[6,1]*100 );
print( conf );
```

Aqui se aplica PCA sobre las características originales. En la ultima parte del código se comenta una sección que es la iteración sobre todos los posibles ejes principales, esta demora mucho por lo que se deja comentada

```python
import numpy as np
import scipy as sp
from sklearn.neural_network import MLPClassifier
#from sklearn import svm
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from matplotlib.mlab import PCA



#funcion para medir desempeños
def TVFP(Confusion):
    #se crea arreglo con 0 para alberga TVP y TFP de cada clase
    TVFP=np.zeros((Confusion.shape[0],2))
    #suma sobre elementos de diagonal
    suma=0
    #Se itera sobre las clases de la matriz de confusion, par obtener TVP y TFP de cada
    #una
    for i in range(0, Confusion.shape[0]):
        #TVP se calcula dividiendo la diagonal de cada clase con el numero total
        #muestras de esa clase
        TVP= Confusion[i,i]/np.sum(Confusion[i,:])
        #TFP se calcula sumando los valores de la clase predicha de la clase y restandole el elemento diagonal,
        #luego esto se divide por todos los datos, menos la fila de la clase
        TFP=(np.sum(Confusion[:,i])-Confusion[i,i])/(np.sum(Confusion)-np.sum(Confusion[i,:]))
        TVFP[i,:]=[TVP,TFP]
        #suma diagonales
        suma=suma+Confusion[i,i]
    
    #Al final se agrega el promedio de TVP y TFP de la red   
    PromTVP=np.sum(TVFP[:,0])/TVFP.shape[0]
    PromTFP=np.sum(TVFP[:,1])/TVFP.shape[0]
    Prom=np.array([[PromTVP,PromTFP]])
    TVFP=np.concatenate((TVFP,Prom), axis=0)
    #se retorna el arreglo
    Acc=suma/np.sum(Confusion)
    return TVFP, Acc

#se carga conjunto de entrenamiento y validacion
Train=np.load('Train.npy')
Val=np.load('Validation.npy') 


#Se separa cada conjunto en la señal y sus clases
sign_train = Train[:,0:6000];
class_train = Train[:,6000];

sign_val = Val[:,0:6000];
class_val = Val[:,6000];


#se caragan las imf previamente calculadas sobre los conjuntos de train y validacion
imf_tr=np.load('imf_tr.npy')
imf_val=np.load('imf_val.npy')       

#se genera un arreglo para registrar accuracy de cada conjunto de caracteristicas
Classifiers=np.zeros((4,2))


#%%

#Clasificacion con caraceristicas originales como en seccion anterior

#funciones de extraccion de caracteristicas entregadas  
def crosses(my_array):
  return ((my_array[:,:-1] * my_array[:,1:]) < 0).sum(axis=1)

def slope_sign_changes(my_array):
    w = np.shape(my_array)[1];
    d1 = my_array[:,1:w-2] - my_array[:,2:w-1]
    d2 = my_array[:,2:w-1] - my_array[:,3:w]
    return np.double( np.sum((d1*d2) < 0, 1) );

def extr_feat(s):
    h = np.shape(s)[0]
    w = np.shape(s)[1];
    feat = np.zeros((h, 10))
    feat[:,0] = np.sum(np.abs(s),1); # iemg
    feat[:,1] = crosses(s); #zc
    feat[:,2] = slope_sign_changes(s); #ssc
    feat[:,3] = np.sum(np.abs(s[:,1:w-1] - s[:,2:w]),1); # wl
    feat[:,4] = np.sum(np.abs(s[:,1:w-1] - s[:,2:w] > 0.1),1); # wamp
    feat[:,5] = np.var(s,1);
    feat[:,6] = sp.stats.skew(s,1);
    feat[:,7] = sp.stats.kurtosis(s,1);
    feat[:,8] = np.median(s,1);
    feat[:,9] = np.std(s,1);
    return feat;

#se extraen las caracteristicas para Train y Validation
caract_train = np.hstack((extr_feat(sign_train[:,1:3000]),
                          extr_feat(sign_train[:,3000:6000]),
                          extr_feat(imf_tr[:,:,0]),
                          extr_feat(imf_tr[:,:,1]),
                          extr_feat(imf_tr[:,:,2]),
                          extr_feat(imf_tr[:,:,3]),
                          extr_feat(imf_tr[:,:,4]),
                          extr_feat(imf_tr[:,:,5])
                         ));

caract_val = np.hstack((extr_feat(sign_val[:,1:3000]),
                         extr_feat(sign_val[:,3000:6000]),
                         extr_feat(imf_val[:,:,0]),
                         extr_feat(imf_val[:,:,1]),
                         extr_feat(imf_val[:,:,2]),
                         extr_feat(imf_val[:,:,3]),
                         extr_feat(imf_val[:,:,4]),
                         extr_feat(imf_val[:,:,5])
                        ));

#se normalizan caracteristicas de cada conjunto, con los parametros del conjunto de train
scaler = preprocessing.StandardScaler().fit(caract_train);
caract_train = scaler.transform(caract_train);
caract_val = scaler.transform(caract_val);



#Clasificacion con MLP
clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300, 50), max_iter = 10000, random_state=1);
clf.fit(caract_train, class_train);
pred = clf.predict(caract_val);
conf = confusion_matrix( class_val, pred);
Rates, Acc= TVFP(conf)
Classifiers[0,0]= Acc*100 

#Clasificacion con RF
clf1 = RandomForestClassifier(n_estimators=50, max_leaf_nodes=100, n_jobs=-1, random_state=0)
clf.fit(caract_train, class_train);
clf1.fit(caract_train, class_train);
pred = clf1.predict(caract_val);
conf = confusion_matrix( class_val, pred);
Rates, Acc= TVFP(conf)
Classifiers[0,1]= Acc*100 



#Se hace PCA sobre las caracteristicas originales
result = PCA(caract_train)
#se calcula varianza a lo largo de componentes principales
PCA_var=result.fracs


#Se proyecta el conjunto de Train en los ejes principales
PCA_train=result.Y
#Se proyecta el conjunto de Validacion en los ejes principales
PCA_val=result.project(caract_val)


#Clasificacion con MLP
clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300, 50), max_iter = 10000, random_state=1);
clf.fit(PCA_train, class_train);
pred = clf.predict(PCA_val);
conf = confusion_matrix( class_val, pred);
Rates, Acc= TVFP(conf)
Classifiers[1,0]= Acc*100 

#Clasificacion con RF
clf = RandomForestClassifier(n_estimators=50, max_leaf_nodes=100, n_jobs=-1, random_state=0)
clf.fit(caract_train, class_train);
clf.fit(PCA_train, class_train);
pred = clf.predict(PCA_val);
conf = confusion_matrix( class_val, pred);
Rates, Acc= TVFP(conf)
Classifiers[1,1]= Acc*100 



#%%
#mejor PCA


#se escogen los primeros 65 ejes que arrojaron el mejor accuracy en Best_PCA sobre la MLP utilizada
#55 ejes principales que contienen el 99.96% de la varianza de los datos

#Se proyecta el conjunto de Train en los ejes principales, tomando las primeras 50 dimensiones
PCA_train=result.Y[:,0:65]
#Se proyecta el conjunto de Validacion en los ejes principales, tomando las primeras 50 dimensiones
PCA_val=result.project(caract_val)[:,0:65]


#Clasificacion con MLP
clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300, 50), max_iter = 10000, random_state=1);
clf.fit(PCA_train, class_train);
pred = clf.predict(PCA_val);
conf = confusion_matrix( class_val, pred);
Rates, Acc= TVFP(conf)
Classifiers[2,0]= Acc*100 

#Clasificacion con RF
clf = RandomForestClassifier(n_estimators=50, max_leaf_nodes=100, n_jobs=-1, random_state=0)
clf.fit(caract_train, class_train);
clf.fit(PCA_train, class_train);
pred = clf.predict(PCA_val);
conf = confusion_matrix( class_val, pred);
Rates, Acc= TVFP(conf)
Classifiers[2,1]= Acc*100 


#%%
#PCA 10 dim

#Se proyecta el conjunto de Train en los ejes principales, tomando las primeras 10 dimensiones
PCA_train=result.Y[:,0:10]
#Se proyecta el conjunto de Validacion en los ejes principales, tomando las primeras 10 dimensiones
PCA_val=result.project(caract_val)[:,0:10]


#Clasificacion con MLP
clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300, 50), max_iter = 10000, random_state=1);
clf.fit(PCA_train, class_train);
pred = clf.predict(PCA_val);
conf = confusion_matrix( class_val, pred);
Rates, Acc= TVFP(conf)
Classifiers[3,0]= Acc*100 

#Clasificacion con RF
clf = RandomForestClassifier(n_estimators=50, max_leaf_nodes=100, n_jobs=-1, random_state=0)
clf.fit(caract_train, class_train);
clf.fit(PCA_train, class_train);
pred = clf.predict(PCA_val);
conf = confusion_matrix( class_val, pred);
Rates, Acc= TVFP(conf)
Classifiers[3,1]= Acc*100 


print('Arreglo de accuracy, primera columna MLP, segunda RF')

print(Classifiers)


##%%
##iteracion sobre todos los ejes principales
#
##se crea lista que contiene desempeño para dimensiones de PCA
#Accuracies=[]
#
#clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300,50), max_iter = 10000, random_state=1);
#
##Feats after PCA
#results = PCA(caract_train)
#PCA_train=results.Y
#PCA_val=results.project(caract_val)
#
#l=caract_train.shape[1]
#for i in range (0, caract_train.shape[1]-1):
#    
#    caract_train = PCA_train[:,0:(l-i)]
#    caract_val = PCA_val[:,0:(l-i)]
#    clf.fit(caract_train, class_train);
#    pred = clf.predict(caract_val);
#    conf = confusion_matrix( class_val, pred);
#    Rates, Acc= TVFP(conf)
#    Accuracies.append(Acc)
##%%
#best=np.asarray(Accuracies)
#
#print(np.where(best==np.max(best)))
##np.save('PCA_best.npy', best)
```

```python
import csv
import numpy as np
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import sklearn as skl 

from sklearn.neural_network import MLPClassifier
from sklearn import svm
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from matplotlib.mlab import PCA
import emd0


#funcion para medir desempeños
def TVFP(Confusion):
    #se crea arreglo con 0 para alberga TVP y TFP de cada clase
    TVFP=np.zeros((Confusion.shape[0],2))
    #suma sobre elementos de diagonal
    suma=0
    #Se itera sobre las clases de la matriz de confusion, par obtener TVP y TFP de cada
    #una
    for i in range(0, Confusion.shape[0]):
        #TVP se calcula dividiendo la diagonal de cada clase con el numero total
        #muestras de esa clase
        TVP= Confusion[i,i]/np.sum(Confusion[i,:])
        #TFP se calcula sumando los valores de la clase predicha de la clase y restandole el elemento diagonal,
        #luego esto se divide por todos los datos, menos la fila de la clase
        TFP=(np.sum(Confusion[:,i])-Confusion[i,i])/(np.sum(Confusion)-np.sum(Confusion[i,:]))
        TVFP[i,:]=[TVP,TFP]
        #suma diagonales
        suma=suma+Confusion[i,i]
    
    #Al final se agrega el promedio de TVP y TFP de la red   
    PromTVP=np.sum(TVFP[:,0])/TVFP.shape[0]
    PromTFP=np.sum(TVFP[:,1])/TVFP.shape[0]
    Prom=np.array([[PromTVP,PromTFP]])
    TVFP=np.concatenate((TVFP,Prom), axis=0)
    #se retorna el arreglo
    Acc=suma/np.sum(Confusion)
    return TVFP, Acc

#se carga conjunto de entrenamiento y validacion
Train=np.load('Train.npy')
Val=np.load('Validation.npy') 


#Se separa cada conjunto en la señal y sus clases
sign_train = Train[:,0:6000];
class_train = Train[:,6000];

sign_val = Val[:,0:6000];
class_val = Val[:,6000];





#%%

#Se separa las señales puras en ventanas de tamaño 300-600-1500-3000 sin traslape
#se tomaran las muestras con ambos canales concatenados

#funcion que recibe un conjunto y el tamaño de una ventana, para retornar
#el conjunto separado en las ventanas
def window_sign(l, Set):
    #dimensiones del conjunto
    shape=Set.shape
    #largo de un canal de las señales del conjunto
    signal_lenght=int((shape[1]-1)/2)
    #cantidad de veces que cabe la ventana en un canal
    steps=  int(signal_lenght/l)
    #arreglo que contendra datos separados en ventanas
    signal=np.zeros((shape[0]*steps,int((shape[1]-1)/steps)))
    #clases de los datos del conjunto
    cls=Set[:,6000];
    #se genera un arreglo de las clases de las ventanas, concatenando cls steps
    #veces; la segunda ventana sucede a la primera en el arreglo de salida
    #manteniendose el orden de las clases en ls conjuntos de cada ventana
    for i in range(0,steps-1):
        cls=np.concatenate((cls, Set[:,6000]), axis=0)

    # se concatenan uno detras de otros los conjuntos de ventanas
    for i in range(0,steps):
        # aqui se concatena las respecti vasprimeras, segundas, ... n, ventanas 
        #de cada canal
        signal[shape[0]*i:shape[0]*(i+1),:]=np.concatenate((Set[0:shape[0],l*i:l*(i+1)],Set[0:shape[0],signal_lenght+(l*i):signal_lenght+(l*(i+1))]), axis=1)
    #se retorna el arreglo de datos y clases
    return signal, cls

#listas que contendran a los conjuntos con distinto largo de ventanas
signw_train=[]
clsw_train=[]

signw_val=[]
clsw_val=[]

#lista de tamaño de ventanas
windows=[300,600,1500,3000]

#se llena las listas de diferentes tamaños de ventanas para cada conjunto
for i in range (0, 4):
    sign_trainw, class_trainw=window_sign(windows[i],Train)
    signw_train.append(sign_trainw)
    clsw_train.append(class_trainw)
    
    sign_valw, class_valw=window_sign(windows[i],Val)
    signw_val.append(sign_valw)
    clsw_val.append(class_valw)
    

#%%
##calculo real de IMF
def IMFs(sign_train):
    htr = sign_train.shape[0];
    l=int(sign_train.shape[1]/2)
    imf_tr = np.zeros((htr,l,6));
    for i in range(htr):
        outimf = emd0.emd0(sign_train[i,0:l],2);
        imf_tr[i,:,0] = outimf[0,:];
        imf_tr[i,:,1] = outimf[1,:];
        imf_tr[i,:,2] = outimf[2,:];
        outimf = emd0.emd0(sign_train[i,l:2*l],2);
        imf_tr[i,:,3] = outimf[0,:];
        imf_tr[i,:,4] = outimf[1,:];
        imf_tr[i,:,5] = outimf[2,:];
        if i % 50 == 0:
            print(i);
    return imf_tr


#imf_train_window=[]
#imf_val_window=[]
#
#
#for i in range (0, 4):
#    imf_train_w=IMFs(signw_train[i])
#    imf_train_window.append(imf_train_w)
#
#    imf_val_w=IMFs(signw_val[i])
#    imf_val_window.append(imf_val_w)
#
#
#imf_train_ventanas=np.asarray(imf_train_window)
#np.save('imf_train_ventanas.npy', imf_train_ventanas)

#imf_val_ventanas=np.asarray(imf_val_window)
#np.save('imf_val_ventanas.npy', imf_val_ventanas)
  
#%%

#se caragan las imf previamente calculadas sobre los conjuntos de train y validacion
imf_tr=np.load('imf_tr.npy')
imf_val=np.load('imf_val.npy')       

#calculo de ventanas sobre IMFs
def window_imf(l, Set):

    shape=Set.shape  
    steps=  int(shape[1]/l)
    signal=np.zeros((shape[0]*steps,int(shape[1]/steps),shape[2]))

    for i in range(0,steps):
        signal[shape[0]*i:shape[0]*(i+1),:,:]=Set[0:shape[0],l*i:l*(i+1),:]

    return signal

#listass que contendran conjuntos de ventanas de IMFs para cada conjunto
imfw_train=[]

imfw_val=[]

#iteracion sobre distintos largos de ventanas
for i in range (0, 4):
    imf_trainw=window_imf(windows[i],imf_tr)
    imfw_train.append(imf_trainw)
        
    imf_valw=window_imf(windows[i],imf_val)
    imfw_val.append(imf_valw)

#%%

# se extraen las características para las ventanas de IMFs
     
def crosses(my_array):
  return ((my_array[:,:-1] * my_array[:,1:]) < 0).sum(axis=1)

def slope_sign_changes(my_array):
    w = np.shape(my_array)[1];
    d1 = my_array[:,1:w-2] - my_array[:,2:w-1]
    d2 = my_array[:,2:w-1] - my_array[:,3:w]
    return np.double( np.sum((d1*d2) < 0, 1) );

def extr_feat(s):
    h = np.shape(s)[0]
    w = np.shape(s)[1];
    feat = np.zeros((h, 10))
    feat[:,0] = np.sum(np.abs(s),1); # iemg_MVA
    feat[:,1] = crosses(s); #zc
    feat[:,2] = slope_sign_changes(s); #ssc
    feat[:,3] = np.sum(np.abs(s[:,1:w-1] - s[:,2:w]),1); # wl
    feat[:,4] = np.sum(np.abs(s[:,1:w-1] - s[:,2:w] > 0.1),1); # wamp
    feat[:,5] = np.var(s,1);
    feat[:,6] = sp.stats.skew(s,1);
    feat[:,7] = sp.stats.kurtosis(s,1);
    feat[:,8] = np.median(s,1);
    feat[:,9] = np.std(s,1);
    return feat;

#lista que contendra características para distintas ventanas de IMFs
feat_train=[]
feat_val=[]

#extraccion de caracteriticas de ventanas de cada IMF
for i in range (0, 4):
    feat_trainw = np.hstack((extr_feat(signw_train[i][:,0:windows[i]]),
                              extr_feat(signw_train[i][:,windows[i]:windows[i]*2]),
                              extr_feat(imfw_train[i][:,:,0]),
                              extr_feat(imfw_train[i][:,:,1]),
                              extr_feat(imfw_train[i][:,:,2]),
                              extr_feat(imfw_train[i][:,:,3]),
                              extr_feat(imfw_train[i][:,:,4]),
                              extr_feat(imfw_train[i][:,:,5])
                             ));
    feat_train.append(feat_trainw)
    
    feat_valw = np.hstack((extr_feat(signw_val[i][:,0:windows[i]]),
                             extr_feat(signw_val[i][:,windows[i]:windows[i]*2]),
                             extr_feat(imfw_val[i][:,:,0]),
                             extr_feat(imfw_val[i][:,:,1]),
                             extr_feat(imfw_val[i][:,:,2]),
                             extr_feat(imfw_val[i][:,:,3]),
                             extr_feat(imfw_val[i][:,:,4]),
                             extr_feat(imfw_val[i][:,:,5])
                            ));
    feat_val.append(feat_valw)


#listas que albergan PCA de ventanas
caract_train=[]
caract_val=[]

#iteracion sobre distintos conjuntos de ventanas
for i in range (0, len(feat_train)):
    results = PCA(feat_train[i])
    PCA_train=results.Y[:,0:10]
    PCA_val=results.project(feat_val[i])[:,0:10]
    caract_train.append(PCA_train)
    caract_val.append(PCA_val)
    

#%%
#se genera un arreglo para registrar accuracy de cada tamaño de ventana
Accuracy=np.zeros(len(feat_train));

class_train=clsw_train
class_val=clsw_val

clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300, 50), max_iter = 10000, random_state=1);
#%%

for i in range (0, len(feat_train)):

    clf.fit(caract_train[i], class_train[i]);
    pred = clf.predict(caract_val[i]);
    conf = confusion_matrix(class_val[i], pred);
    Rates, Acc= TVFP(conf)
    Accuracy[i]= Acc*100 


#%%
print('Arreglo de accuracy, cada elemento una ventana distinta con MLP')

print(Accuracy)
```

```python
import csv
import numpy as np
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import sklearn as skl 

from sklearn.neural_network import MLPClassifier
from sklearn import svm
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from matplotlib.mlab import PCA
import emd0


#funcion para medir desempeños
def TVFP(Confusion):
    #se crea arreglo con 0 para alberga TVP y TFP de cada clase
    TVFP=np.zeros((Confusion.shape[0],2))
    #suma sobre elementos de diagonal
    suma=0
    #Se itera sobre las clases de la matriz de confusion, par obtener TVP y TFP de cada
    #una
    for i in range(0, Confusion.shape[0]):
        #TVP se calcula dividiendo la diagonal de cada clase con el numero total
        #muestras de esa clase
        TVP= Confusion[i,i]/np.sum(Confusion[i,:])
        #TFP se calcula sumando los valores de la clase predicha de la clase y restandole el elemento diagonal,
        #luego esto se divide por todos los datos, menos la fila de la clase
        TFP=(np.sum(Confusion[:,i])-Confusion[i,i])/(np.sum(Confusion)-np.sum(Confusion[i,:]))
        TVFP[i,:]=[TVP,TFP]
        #suma diagonales
        suma=suma+Confusion[i,i]
    
    #Al final se agrega el promedio de TVP y TFP de la red   
    PromTVP=np.sum(TVFP[:,0])/TVFP.shape[0]
    PromTFP=np.sum(TVFP[:,1])/TVFP.shape[0]
    Prom=np.array([[PromTVP,PromTFP]])
    TVFP=np.concatenate((TVFP,Prom), axis=0)
    #se retorna el arreglo
    Acc=suma/np.sum(Confusion)
    return TVFP, Acc

#se carga conjunto de entrenamiento y validacion
Train=np.load('Train.npy')
Val=np.load('Validation.npy') 


#Se separa cada conjunto en la señal y sus clases
sign_train = Train[:,0:6000];
class_train = Train[:,6000];

sign_val = Val[:,0:6000];
class_val = Val[:,6000];





#%%

#Se separa las señales puras en ventanas de tamaño 300-600-1500-3000 con traslape de 50%
#se tomaran las muestras con ambos canales concatenados

#funcion que recibe un conjunto y el tamaño de una ventana, para retornar
#el conjunto separado en las ventanas
def window_sign_overlap(l, Set):
    #dimensiones del conjunto
    shape=Set.shape
    #largo de un canal de las señales del conjunto
    signal_lenght=int((shape[1]-1)/2)
    #cantidad de veces que cabe la ventana en un canal
    steps=  int(signal_lenght/l)
    #arreglo que contendra datos separados en ventanas
    signal=np.zeros((shape[0]*(steps*2-1),int((shape[1]-1)/steps)))
    #clases de los datos del conjunto
    cls=Set[:,6000];
    #se genera un arreglo de las clases de las ventanas, concatenando cls steps
    #veces; la segunda ventana sucede a la primera en el arreglo de salida
    #manteniendose el orden de las clases en ls conjuntos de cada ventana
    for i in range(0,(steps*2-1)-1):
        cls=np.concatenate((cls, Set[:,6000]), axis=0)

    # se concatenan uno detras de otros los conjuntos de ventanas
    for i in range(0,steps):
        # aqui se concatena las respecti vasprimeras, segundas, ... n, ventanas 
        #de cada canal
        signal[shape[0]*i*2:shape[0]*(2*i+1),:]=np.concatenate((Set[0:shape[0],l*i:l*(i+1)],Set[0:shape[0],signal_lenght+(l*i):signal_lenght+(l*(i+1))]), axis=1)
        #condicion que agrega ventana traslapada
        if i!=(steps-1):
            signal[shape[0]*(i+1):shape[0]*(i+2),:]=np.concatenate((Set[0:shape[0],l*(i+1)-int(l/2):l*(i+2)-int(l/2)],Set[0:shape[0],signal_lenght+(l*(i+1)-int(l/2)):signal_lenght+(l*(i+2)-int(l/2))]), axis=1)

    #se retorna el arreglo de datos y clases
    return signal, cls

#listas que contendran a los conjuntos con distinto largo de ventanas
signw_train=[]
clsw_train=[]

signw_val=[]
clsw_val=[]

#lista de tamaño de ventanas
windows=[300,600,1500,3000]

#se llena las listas de diferentes tamaños de ventanas para cada conjunto
for i in range (0, 4):
    sign_trainw, class_trainw=window_sign_overlap(windows[i],Train)
    signw_train.append(sign_trainw)
    clsw_train.append(class_trainw)
    
    sign_valw, class_valw=window_sign_overlap(windows[i],Val)
    signw_val.append(sign_valw)
    clsw_val.append(class_valw)
    

#%%
##calculo real de IMF
def IMFs(sign_train):
    htr = sign_train.shape[0];
    l=int(sign_train.shape[1]/2)
    imf_tr = np.zeros((htr,l,6));
    for i in range(htr):
        outimf = emd0.emd0(sign_train[i,0:l],2);
        imf_tr[i,:,0] = outimf[0,:];
        imf_tr[i,:,1] = outimf[1,:];
        imf_tr[i,:,2] = outimf[2,:];
        outimf = emd0.emd0(sign_train[i,l:2*l],2);
        imf_tr[i,:,3] = outimf[0,:];
        imf_tr[i,:,4] = outimf[1,:];
        imf_tr[i,:,5] = outimf[2,:];
        if i % 50 == 0:
            print(i);
    return imf_tr


#imf_train_window=[]
#imf_val_window=[]
#
#
#for i in range (0, 4):
#    imf_train_w=IMFs(signw_train[i])
#    imf_train_window.append(imf_train_w)
#
#    imf_val_w=IMFs(signw_val[i])
#    imf_val_window.append(imf_val_w)
#
#
#imf_train_ventanas=np.asarray(imf_train_window)
#np.save('imf_train_ventanas.npy', imf_train_ventanas)

#imf_val_ventanas=np.asarray(imf_val_window)
#np.save('imf_val_ventanas.npy', imf_val_ventanas)
  
#%%

#se caragan las imf previamente calculadas sobre los conjuntos de train y validacion
imf_tr=np.load('imf_tr.npy')
imf_val=np.load('imf_val.npy')       

#calculo de ventanas sobre IMFs
def window_imf_overlap(l, Set):
    
    shape=Set.shape  
    steps=  int(shape[1]/l)
    signal=np.zeros((shape[0]*(steps*2-1),int(shape[1]/steps),shape[2]))

    for i in range(0,steps):
        signal[shape[0]*i*2:shape[0]*(i*2+1),:,:]=Set[0:shape[0],l*i:l*(i+1),:]

        if i!=(steps-1):
            signal[shape[0]*(i+1):shape[0]*(i+2),:]=Set[0:shape[0],l*(i+1)-int(l/2):l*(i+2)-int(l/2),:]

    return signal

#listass que contendran conjuntos de ventanas de IMFs para cada conjunto
imfw_train=[]

imfw_val=[]

#iteracion sobre distintos largos de ventanas
for i in range (0, 4):
    imf_trainw=window_imf_overlap(windows[i],imf_tr)
    imfw_train.append(imf_trainw)
        
    imf_valw=window_imf_overlap(windows[i],imf_val)
    imfw_val.append(imf_valw)

#%%

# se extraen las características para las ventanas de IMFs
     
def crosses(my_array):
  return ((my_array[:,:-1] * my_array[:,1:]) < 0).sum(axis=1)

def slope_sign_changes(my_array):
    w = np.shape(my_array)[1];
    d1 = my_array[:,1:w-2] - my_array[:,2:w-1]
    d2 = my_array[:,2:w-1] - my_array[:,3:w]
    return np.double( np.sum((d1*d2) < 0, 1) );

def extr_feat(s):
    h = np.shape(s)[0]
    w = np.shape(s)[1];
    feat = np.zeros((h, 10))
    feat[:,0] = np.sum(np.abs(s),1); # iemg_MVA
    feat[:,1] = crosses(s); #zc
    feat[:,2] = slope_sign_changes(s); #ssc
    feat[:,3] = np.sum(np.abs(s[:,1:w-1] - s[:,2:w]),1); # wl
    feat[:,4] = np.sum(np.abs(s[:,1:w-1] - s[:,2:w] > 0.1),1); # wamp
    feat[:,5] = np.var(s,1);
    feat[:,6] = sp.stats.skew(s,1);
    feat[:,7] = sp.stats.kurtosis(s,1);
    feat[:,8] = np.median(s,1);
    feat[:,9] = np.std(s,1);
    return feat;

#lista que contendra características para distintas ventanas de IMFs
feat_train=[]
feat_val=[]

#extraccion de caracteriticas de ventanas de cada IMF
for i in range (0, 4):
    feat_trainw = np.hstack((extr_feat(signw_train[i][:,0:windows[i]]),
                              extr_feat(signw_train[i][:,windows[i]:windows[i]*2]),
                              extr_feat(imfw_train[i][:,:,0]),
                              extr_feat(imfw_train[i][:,:,1]),
                              extr_feat(imfw_train[i][:,:,2]),
                              extr_feat(imfw_train[i][:,:,3]),
                              extr_feat(imfw_train[i][:,:,4]),
                              extr_feat(imfw_train[i][:,:,5])
                             ));
    feat_train.append(feat_trainw)
    
    feat_valw = np.hstack((extr_feat(signw_val[i][:,0:windows[i]]),
                             extr_feat(signw_val[i][:,windows[i]:windows[i]*2]),
                             extr_feat(imfw_val[i][:,:,0]),
                             extr_feat(imfw_val[i][:,:,1]),
                             extr_feat(imfw_val[i][:,:,2]),
                             extr_feat(imfw_val[i][:,:,3]),
                             extr_feat(imfw_val[i][:,:,4]),
                             extr_feat(imfw_val[i][:,:,5])
                            ));
    feat_val.append(feat_valw)

#calculo de PCA

caract_train=[]
caract_val=[]

#iteracion sobre distintos conjuntos de ventanas
for i in range (0, len(feat_train)):
    results = PCA(feat_train[i])
    PCA_train=results.Y[:,0:10]
    PCA_val=results.project(feat_val[i])[:,0:10]
    caract_train.append(PCA_train)
    caract_val.append(PCA_val)
    

#%%
#se genera un arreglo para registrar accuracy de cada tamaño de ventana
Accuracy=np.zeros(len(feat_train));

class_train=clsw_train
class_val=clsw_val

clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300, 50), max_iter = 10000, random_state=1);
#%%

for i in range (0, len(feat_train)):

    clf.fit(caract_train[i], class_train[i]);
    pred = clf.predict(caract_val[i]);
    conf = confusion_matrix(class_val[i], pred);
    Rates, Acc= TVFP(conf)
    Accuracy[i]= Acc*100 


#%%
print('Arreglo de accuracy, cada elemento una ventana distinta con MLP')

print(Accuracy)
```

```python
import csv
import numpy as np
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import sklearn as skl 

from sklearn.neural_network import MLPClassifier
from sklearn import svm
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from matplotlib.mlab import PCA
import emd0
import itertools

#funcion para medir desempeños
def TVFP(Confusion):
    #se crea arreglo con 0 para alberga TVP y TFP de cada clase
    TVFP=np.zeros((Confusion.shape[0],2))
    #suma sobre elementos de diagonal
    suma=0
    #Se itera sobre las clases de la matriz de confusion, par obtener TVP y TFP de cada
    #una
    for i in range(0, Confusion.shape[0]):
        #TVP se calcula dividiendo la diagonal de cada clase con el numero total
        #muestras de esa clase
        TVP= Confusion[i,i]/np.sum(Confusion[i,:])
        #TFP se calcula sumando los valores de la clase predicha de la clase y restandole el elemento diagonal,
        #luego esto se divide por todos los datos, menos la fila de la clase
        TFP=(np.sum(Confusion[:,i])-Confusion[i,i])/(np.sum(Confusion)-np.sum(Confusion[i,:]))
        TVFP[i,:]=[TVP,TFP]
        #suma diagonales
        suma=suma+Confusion[i,i]
    
    #Al final se agrega el promedio de TVP y TFP de la red   
    PromTVP=np.sum(TVFP[:,0])/TVFP.shape[0]
    PromTFP=np.sum(TVFP[:,1])/TVFP.shape[0]
    Prom=np.array([[PromTVP,PromTFP]])
    TVFP=np.concatenate((TVFP,Prom), axis=0)
    #se retorna el arreglo
    Acc=suma/np.sum(Confusion)
    return TVFP, Acc

#se carga conjunto de entrenamiento y test
Train=np.load('Train.npy')
Test=np.load('Test.npy') 




#se caragan las imf previamente calculadas sobre los conjuntos de train y test
imf_tr=np.load('imf_tr.npy')
imf_te=np.load('imf_te.npy')  

Classifiers=np.zeros((4,2))


#Funcion para mostrar matrices de confusion
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Greens):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=14)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label',fontsize=12)
    plt.xlabel('Predicted label',fontsize=12)
    plt.show()
    
#%%

#Ventana 3000-Caract orig- MLP/RF

#Se separa las señales puras en ventanas de tamaño 300-600-1500-3000 sin traslape
#se tomaran las muestras con ambos canales concatenados

#funcion que recibe un conjunto y el tamaño de una ventana, para retornar
#el conjunto separado en las ventanas
def window_sign(l, Set):
    #dimensiones del conjunto
    shape=Set.shape
    #largo de un canal de las señales del conjunto
    signal_lenght=int((shape[1]-1)/2)
    #cantidad de veces que cabe la ventana en un canal
    steps=  int(signal_lenght/l)
    #arreglo que contendra datos separados en ventanas
    signal=np.zeros((shape[0]*steps,int((shape[1]-1)/steps)))
    #clases de los datos del conjunto
    cls=Set[:,6000];
    #se genera un arreglo de las clases de las ventanas, concatenando cls steps
    #veces; la segunda ventana sucede a la primera en el arreglo de salida
    #manteniendose el orden de las clases en ls conjuntos de cada ventana
    for i in range(0,steps-1):
        cls=np.concatenate((cls, Set[:,6000]), axis=0)

    # se concatenan uno detras de otros los conjuntos de ventanas
    for i in range(0,steps):
        # aqui se concatena las respecti vasprimeras, segundas, ... n, ventanas 
        #de cada canal
        signal[shape[0]*i:shape[0]*(i+1),:]=np.concatenate((Set[0:shape[0],l*i:l*(i+1)],Set[0:shape[0],signal_lenght+(l*i):signal_lenght+(l*(i+1))]), axis=1)
    #se retorna el arreglo de datos y clases
    return signal, cls



sign_train, class_train=window_sign(3000,Train)
sign_test, class_test=window_sign(3000,Test)

     

#calculo de ventanas sobre IMFs
def window_imf(l, Set):

    shape=Set.shape  
    steps=  int(shape[1]/l)
    signal=np.zeros((shape[0]*steps,int(shape[1]/steps),shape[2]))

    for i in range(0,steps):
        signal[shape[0]*i:shape[0]*(i+1),:,:]=Set[0:shape[0],l*i:l*(i+1),:]

    return signal


#iteracion sobre distintos largos de ventanas
imf_train=window_imf(3000,imf_tr)
imf_test=window_imf(3000,imf_te)


# se extraen las características para las ventanas de IMFs
     
def crosses(my_array):
  return ((my_array[:,:-1] * my_array[:,1:]) < 0).sum(axis=1)

def slope_sign_changes(my_array):
    w = np.shape(my_array)[1];
    d1 = my_array[:,1:w-2] - my_array[:,2:w-1]
    d2 = my_array[:,2:w-1] - my_array[:,3:w]
    return np.double( np.sum((d1*d2) < 0, 1) );

def extr_feat(s):
    h = np.shape(s)[0]
    w = np.shape(s)[1];
    feat = np.zeros((h, 10))
    feat[:,0] = np.sum(np.abs(s),1); # iemg_MVA
    feat[:,1] = crosses(s); #zc
    feat[:,2] = slope_sign_changes(s); #ssc
    feat[:,3] = np.sum(np.abs(s[:,1:w-1] - s[:,2:w]),1); # wl
    feat[:,4] = np.sum(np.abs(s[:,1:w-1] - s[:,2:w] > 0.1),1); # wamp
    feat[:,5] = np.var(s,1);
    feat[:,6] = sp.stats.skew(s,1);
    feat[:,7] = sp.stats.kurtosis(s,1);
    feat[:,8] = np.median(s,1);
    feat[:,9] = np.std(s,1);
    return feat;


#extraccion de caracteriticas de ventanas de cada IMF
feat_train = np.hstack((extr_feat(sign_train[:,0:3000]),
                              extr_feat(sign_train[:,3000:3000*2]),
                              extr_feat(imf_train[:,:,0]),
                              extr_feat(imf_train[:,:,1]),
                              extr_feat(imf_train[:,:,2]),
                              extr_feat(imf_train[:,:,3]),
                              extr_feat(imf_train[:,:,4]),
                              extr_feat(imf_train[:,:,5])
                             ));

    
feat_test = np.hstack((extr_feat(sign_test[:,0:3000]),
                             extr_feat(sign_test[:,3000:3000*2]),
                             extr_feat(imf_test[:,:,0]),
                             extr_feat(imf_test[:,:,1]),
                             extr_feat(imf_test[:,:,2]),
                             extr_feat(imf_test[:,:,3]),
                             extr_feat(imf_test[:,:,4]),
                             extr_feat(imf_test[:,:,5])
                            ));



#PCA

#se normalizan caracteristicas de cada conjunto, con los parametros del conjunto de train
scaler = preprocessing.StandardScaler().fit(feat_train);
Orig_train= scaler.transform(feat_train);
Orig_test = scaler.transform(feat_test);


#Clasificacion con MLP
clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300, 50), max_iter = 10000, random_state=1);
clf.fit(Orig_train, class_train);
pred = clf.predict(Orig_test);
conf_orig_MLP = confusion_matrix( class_test, pred);
Rates, Acc= TVFP(conf_orig_MLP)
Classifiers[0,0]= Acc*100 

print('Acc= ', Acc*100)
cls=['0','1','2','3','4','5']
fig=plt.figure()
plot_confusion_matrix(conf_orig_MLP, classes=cls,title='No Window-Orig-MLP')
#fig.savefig("CM_orig_MLP.png",bbox_inches='tight')

#Clasificacion con RF
clf = RandomForestClassifier(n_estimators=50, max_leaf_nodes=100, n_jobs=-1, random_state=0)
clf.fit(Orig_train, class_train);
pred = clf.predict(Orig_test);
conf_orig_RF = confusion_matrix( class_test, pred);
Rates, Acc= TVFP(conf_orig_RF)
Classifiers[0,1]= Acc*100 

print('Acc= ', Acc*100)
cls=['0','1','2','3','4','5']
fig=plt.figure()
plot_confusion_matrix(conf_orig_RF, classes=cls,title='No Window-PCA-RF')
#fig.savefig("CM_orig_RF.png",bbox_inches='tight')

#%%

#Ventana 3000-PCA 10- MLP/RF

results = PCA(feat_train)
PCA_train_10=results.Y[:,0:10]
PCA_test_10=results.project(feat_test)[:,0:10]


#Clasificacion con MLP
clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300, 50), max_iter = 10000, random_state=1);
clf.fit(PCA_train_10, class_train);
pred = clf.predict(PCA_test_10);
conf_10_MLP = confusion_matrix( class_test, pred);
Rates, Acc= TVFP(conf_10_MLP)
Classifiers[1,0]= Acc*100 

print('Acc= ', Acc*100)
cls=['0','1','2','3','4','5']
fig=plt.figure()
plot_confusion_matrix(conf_10_MLP, classes=cls,title='No Window-PCA-MLP')
#fig.savefig("CM_PCA_MLP.png",bbox_inches='tight')

#Clasificacion con RF
clf = RandomForestClassifier(n_estimators=50, max_leaf_nodes=100, n_jobs=-1, random_state=0)
clf.fit(PCA_train_10, class_train);
pred = clf.predict(PCA_test_10);
conf_10_RF = confusion_matrix( class_test, pred);
Rates, Acc= TVFP(conf_10_RF)
Classifiers[1,1]= Acc*100 

print('Acc= ', Acc*100)
cls=['0','1','2','3','4','5']
fig=plt.figure()
plot_confusion_matrix(conf_10_RF, classes=cls,title='No Window-PCA-RF')
#fig.savefig("CM_PCA_RF.png",bbox_inches='tight')



#%%

#Ventana 1500-Caract orig- MLP/RF



signw_train, class_trainw=window_sign(1500,Train)
signw_test, class_testw=window_sign(1500,Test)

     

#calculo de ventanas sobre IMFs
def window_imf(l, Set):

    shape=Set.shape  
    steps=  int(shape[1]/l)
    signal=np.zeros((shape[0]*steps,int(shape[1]/steps),shape[2]))

    for i in range(0,steps):
        signal[shape[0]*i:shape[0]*(i+1),:,:]=Set[0:shape[0],l*i:l*(i+1),:]

    return signal


#iteracion sobre distintos largos de ventanas
imfw_train=window_imf(1500,imf_tr)
imfw_test=window_imf(1500,imf_te)


# se extraen las características para las ventanas de IMFs
     


#extraccion de caracteriticas de ventanas de cada IMF
feat_trainw = np.hstack((extr_feat(signw_train[:,0:1500]),
                              extr_feat(signw_train[:,1500:1500*2]),
                              extr_feat(imfw_train[:,:,0]),
                              extr_feat(imfw_train[:,:,1]),
                              extr_feat(imfw_train[:,:,2]),
                              extr_feat(imfw_train[:,:,3]),
                              extr_feat(imfw_train[:,:,4]),
                              extr_feat(imfw_train[:,:,5])
                             ));

    
feat_testw = np.hstack((extr_feat(signw_test[:,0:1500]),
                             extr_feat(signw_test[:,1500:1500*2]),
                             extr_feat(imfw_test[:,:,0]),
                             extr_feat(imfw_test[:,:,1]),
                             extr_feat(imfw_test[:,:,2]),
                             extr_feat(imfw_test[:,:,3]),
                             extr_feat(imfw_test[:,:,4]),
                             extr_feat(imfw_test[:,:,5])
                            ));



#PCA

#se normalizan caracteristicas de cada conjunto, con los parametros del conjunto de train
scaler = preprocessing.StandardScaler().fit(feat_trainw);
Orig_train_w= scaler.transform(feat_trainw);
Orig_test_w = scaler.transform(feat_testw);


#Clasificacion con MLP
clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300, 50), max_iter = 10000, random_state=1);
clf.fit(Orig_train_w, class_trainw);
pred = clf.predict(Orig_test_w);
conf_w_orig_MLP = confusion_matrix( class_testw, pred);
Rates, Acc= TVFP(conf_w_orig_MLP)
Classifiers[2,0]= Acc*100 
print('Acc= ', Acc*100)

cls=['0','1','2','3','4','5']
fig=plt.figure()
plot_confusion_matrix(conf_w_orig_MLP, classes=cls,title='Window-Orig-MLP')
#fig.savefig("CM_w_orig_MLP.png",bbox_inches='tight')

#Clasificacion con RF
clf = RandomForestClassifier(n_estimators=50, max_leaf_nodes=100, n_jobs=-1, random_state=0)
clf.fit(Orig_train_w, class_trainw);
pred = clf.predict(Orig_test_w);
conf_w_10_RF = confusion_matrix( class_testw, pred);
Rates, Acc= TVFP(conf_w_10_RF)
Classifiers[2,1]= Acc*100 

print('Acc= ', Acc*100)
cls=['0','1','2','3','4','5']
fig=plt.figure()
plot_confusion_matrix(conf_w_10_RF, classes=cls,title='Window-PCA-RF')
#fig.savefig("CM_w_orig_RF.png",bbox_inches='tight')



#%%

#Ventana 1500-PCA 10- MLP/RF

results_w = PCA(feat_trainw)
PCA_train_w_10=results_w.Y[:,0:10]
PCA_test_w_10=results_w.project(feat_testw)[:,0:10]


#Clasificacion con MLP
clf = MLPClassifier(solver='adam', alpha=1e-5, tol=1e-5, hidden_layer_sizes=(300, 50), max_iter = 10000, random_state=1);
clf.fit(PCA_train_w_10, class_trainw);
pred = clf.predict(PCA_test_w_10);
conf_w_10_MLP = confusion_matrix( class_testw, pred);
Rates, Acc= TVFP(conf_w_10_MLP)
Classifiers[3,0]= Acc*100 
print('Acc= ', Acc*100)

cls=['0','1','2','3','4','5']
fig=plt.figure()
plot_confusion_matrix(conf_w_10_MLP, classes=cls,title='Window-PCA-MLP')
#fig.savefig("CM_w_PCA_MLP.png",bbox_inches='tight')

#Clasificacion con RF
clf = RandomForestClassifier(n_estimators=50, max_leaf_nodes=100, n_jobs=-1, random_state=0)
clf.fit(PCA_train_w_10, class_trainw);
pred = clf.predict(PCA_test_w_10);
conf_w_10_RF = confusion_matrix( class_testw, pred);
Rates, Acc= TVFP(conf_w_10_RF)
Classifiers[3,1]= Acc*100 
print('Acc= ', Acc*100)

cls=['0','1','2','3','4','5']
fig=plt.figure()
plot_confusion_matrix(conf_w_10_RF, classes=cls,title='Window-PCA-RF')
#fig.savefig("CM_w_PCA_RF.png",bbox_inches='tight')
```
